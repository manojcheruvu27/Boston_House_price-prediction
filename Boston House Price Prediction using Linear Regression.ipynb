# Linear regression with SciKit-Learn

```python
df.head()
```

### As we have seen RM has huge impact on MEDV(Less correlated so we use RM as independent variable and MEDV as dependent)

```python
X = df['RM'].values.reshape(-1,1)
```

```python
Y = df['MEDV'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()      ##This is one instance of model if we want to apply linear regression on other models then we can say 
                                ## m1,m2..= LinearRegression()
                                ## m1.fit(a,b), m2.fit(p,q)...
```

```python
model.fit(X,Y)
```

```python
LinearRegression(copy_X=True,fit_intercept = True, n_jobs=None, normalize=False)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(12,10))
sns.regplot(X,Y)
plt.xlabel('Average no of rooms per dwelling')
plt.ylabel("Median values of owner occupied homes in $1000's")
plt.show()
```

```python
sns.jointplot(x='RM',y='MEDV',data=df,kind='reg',height=8)
```
